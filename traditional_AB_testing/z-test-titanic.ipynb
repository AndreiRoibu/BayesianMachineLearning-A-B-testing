```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from statsmodels.stats.weightstats import ztest
```

```python
file_name = '../../large_files/titanic_train.csv'
```

```python
df = pd.read_csv(file_name)
```

```python
df.head()
```

```python
df[df['Survived']==1].head()
```

```python
x1 = df[df['Survived']==1]['Fare'].dropna().to_numpy()
```

```python
x2 = df[df['Survived']==0]['Fare'].dropna().to_numpy()
```

```python
sns.kdeplot(x1, label='survived')
sns.kdeplot(x2, label='did not survive')
plt.legend()
```

```python
print(x1.mean(), x2.mean()) # we can see that there is a differnece between the two groups! higher fares = better survival chances!
```

```python
ztest(x1, x2)
```
